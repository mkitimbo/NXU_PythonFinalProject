```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For Jupyter Notebook, use current working directory
BASE_DIR = os.getcwd()
DATA_DIR = os.path.join(BASE_DIR, "data")
```

```python
# Make sure the data folder exists
os.makedirs(DATA_DIR, exist_ok=True)
```

```python
# Load CSV file
csv_path = os.path.join(DATA_DIR, "users.csv")
df = pd.read_csv(csv_path)
```

```python
#Create and save chart: Age vs Income
plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='income', data=df.sort_values('income', ascending=False))
plt.title('Ages with Highest Income')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Save chart to the correct path
chart_path = os.path.join(DATA_DIR, "age_income_chart.png")
plt.savefig(chart_path)
plt.show()
```

```python
# Gender Distribution Across Spending Categories
expenses = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
df_melted = df.melt(id_vars=['gender'], value_vars=expenses, var_name='category', value_name='amount')

plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='amount', hue='gender', data=df_melted)
plt.title('Gender Distribution Across Expense Categories')
plt.tight_layout()

# Save the chart to the correct path
chart_path = os.path.join(DATA_DIR, "gender_spending_chart.png")
plt.savefig(chart_path)
plt.show()
```
